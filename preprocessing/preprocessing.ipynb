```python
# import libraries

import pandas as pd
import numpy as np
```

### Satisfaction dataset

```python
data = pd.read_csv('../dataset/istat_soddisfazione.csv')

regioni = ['Italia','Trentino Alto Adige / Südtirol', 'Emilia-Romagna','Toscana','Abruzzo','Piemonte','Lazio', 'Sardegna', 'Calabria','Lombardia', 'Sicilia','Basilicata', 'Puglia', 'Campania','Provincia Autonoma Bolzano / Bozen','Provincia Autonoma Trento', 'Friuli-Venezia Giulia',"Valle d'Aosta / Vallée d'Aoste",'Veneto', 'Marche','Liguria', 'Umbria', 'Molise']
data = data.loc[data['Territorio'].isin(regioni)]
for index, row in data.iterrows():
    if row['TIPO_DATO_AVQ'] in ['0_HFOOD_Q','0_HFOOD_V']:
        data.loc[index,'TIPO_ANALISI'] = 'FOOD'
    if row['TIPO_DATO_AVQ'] in ['0_SANFAC_V','0_SANFAC_Q']:
        data.loc[index,'TIPO_ANALISI'] = 'SANFAC'
    if row['TIPO_DATO_AVQ'] in ['0_NURS_V','0_NURS_Q']:
        data.loc[index,'TIPO_ANALISI'] = 'NURS'
    if row['TIPO_DATO_AVQ'] in ['0_MED_CAREQ','0_MED_CAREV']:
        data.loc[index,'TIPO_ANALISI'] = 'MED_CARE'
data = data[['ITTER107','Territorio','TIPO_ANALISI','Tipo dato','MISURA_AVQ','Misura','TIME','Value']]


for index, row in data.iterrows():
    if row['Tipo dato'] == 'molto ':
        data.loc[index,'value_molto'] = row['Value']
    if row['Tipo dato'] == 'molto e abbastanza':
        data.loc[index,'value_molto&abbastanza'] = row['Value']

del data['Tipo dato']
df = pd.DataFrame()
for index, row in data.iterrows():
    
    if  not (pd.isnull(row['value_molto&abbastanza'])) :
        d = data.loc[data['TIPO_ANALISI'] == row['TIPO_ANALISI']].loc[data['Territorio'] == row['Territorio']].loc[data['MISURA_AVQ'] == row['MISURA_AVQ']].loc[data['TIME'] == row['TIME']].loc[data['value_molto&abbastanza'].isnull()]
        d['value_molto&abbastanza'] = row['value_molto&abbastanza']
        
        df = df.append(d)
    if  not (pd.isnull(row['value_molto'])) :
        b = data.loc[data['TIPO_ANALISI'] == row['TIPO_ANALISI']].loc[data['Territorio'] == row['Territorio']].loc[data['MISURA_AVQ'] == row['MISURA_AVQ']].loc[data['TIME'] == row['TIME']].loc[data['value_molto'].isnull()]
        b['value_molto'] = row['value_molto']
        
        df = df.append(b)
del df['Value']

df = df.drop_duplicates()

df = df[['Territorio','TIPO_ANALISI','MISURA_AVQ','TIME','value_molto&abbastanza','value_molto']]
df['value_abbastanza'] = df['value_molto&abbastanza'] - df['value_molto']
df['value_abbastanza'] = round(df['value_abbastanza'],1)
df = df.loc[df['MISURA_AVQ'] == 'HSC']
df['value_scarso&poco'] = 100 - df['value_molto&abbastanza']
del df['value_molto&abbastanza']
df['value_scarso&poco'] = round(df['value_scarso&poco'],1)
df
```

```python
soddisfazione = df
```

```python
soddisfazione
```

```python
soddisfazione.to_csv('../dataset/clean_dataset/soddisfazione.csv')
```
